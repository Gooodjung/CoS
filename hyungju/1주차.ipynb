####2753. 윤년

```python
year = int(input()) #출력 시 숫자 입력

if year%4==0 and year%100 !=0 or year%400==0: #if문에서 조건 여러개 작성 가능
    print(1)
else:
    print(0)
```

####2750. 수 정렬하기

```python
N = int(input())
m = [] #리스트 생성

#뭐든 리스트를 세로로 나열하려면 for문 쓰기
for i in range(N):
    m.append(int(input()))

for i in range(len(m)):
    for j in range(len(m)):
        if m[i]<m[j]:
            m[i],m[j] = m[j],m[i] #스와프 (리스트에서 요소 순서 바꾸는 - 3개도 가능)

#그냥 print(m)하면 [1,2,3,4,5] 리스트 형태로 출력, 세로로 나열하려면 for문
for k in m:
    print(k)
```

####10989. 수 정렬하기3

```python
N = int(input())
m=[]

for i in range(N):
    m.append(int(input()))

for i in range(len(m)):
    for j in range(len(m)):
        if m[i] < m[j]:
            m[i], m[j] = m[j], m[i]
        if m[i]==m[j]:
            m[i], m[j] = m[i], m[j]

for k in m:
    print(k)
```

####2605. 줄 세우기

```python
N = int(input())
num = list(map(int,input().split()))
line=[]

for i in range(N):
    line.insert(i-num[i],i+1)

print(line)
```

####2884. 알람 시계

```python
H, M = input().split()
H = int(H)
M = int(M)

if H ==0 and M < 45:
    print(23, M+15)
elif M < 45: #elif문 사용해서 0시일 때만 따로 조건 부여
    print(H-1, M+15)
else:
    print(H,M-45)
```

####1110. 더하기 사이클

```python
N=int(input())
num = N
iter=0

#일의 자릿수, 십의 자릿수는 //, % (나누기) 연산자로 구하기
while True: #infinite loop 시행, 원래 숫자로 돌아올 때까지 시행해야 하니까
    a = num%10 #일의 자리
    b = ((num//10)+(num%10))%10 
    num= 10*a + b #num을 계속 갱신
    iter=iter+1 #시행 횟수 

    if num == N: #원래 숫자와 같아지면
        break #loop 중단 

print(iter)
```
